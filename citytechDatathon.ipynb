```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('Demograph.csv')
```

```python
df.head()
```

```python
df_small = df[:5]
df_small
```

```python
conn = sqlite3.connect(':memory:')  # Use ':memory:' to create a database in RAM
```

```python
df_small.to_sql('records', conn, index=False, if_exists='replace')
```

```python
df.to_sql('recordsAll', conn, index=False, if_exists='replace')
```

```python
df.rename(columns={
    'Data Year': 'year',
    'Agency Name': 'agency',
    'Equal Employment Opportunity (EEO)-4 Job Category': 'EEO',
    'Upper Pay Band Bound': 'upperpay',
    'Number of Employees': 'numEmp',
    'Lower Pay Band Bound': 'lowerpay'
}, inplace=True)

df.head()
```

```python
query = "SELECT * FROM recordsAll WHERE Gender = 'Male'"
result_df = pd.read_sql_query(query, conn)

# Display the result
result_df
```

```python
query_2 = 'SELECT "Data Year", "Gender", "Race", AVG("Upper Pay Band Bound") AS avg_upper_pay FROM recordsAll GROUP BY "Data Year", "Gender", "Race" ORDER BY  "Data Year", "Race", "Gender" ASC';
result_df_2 = pd.read_sql_query(query_2, conn)
# Display the result
result_df_2
```

```python
query_3 = 'SELECT "Equal Employment Opportunity (EEO)-4 Job Category", "Employee Status", COUNT(*) FROM recordsAll WHERE "Race" = "Black or African American" GROUP BY "Equal Employment Opportunity (EEO)-4 Job Category", "Employee Status" ORDER BY  COUNT(*) ASC'
result_df_3 = pd.read_sql_query(query_3, conn)
# Display the result
result_df_3
```

```python
query_4 = 'SELECT COUNT(*) FROM recordsAll WHERE "Ethnicity" = "Hispanic or Latino"'
result_df_4 = pd.read_sql_query(query_4, conn)
# Display the result
result_df_4


```

```python
# Close the connection
conn.close()
```
